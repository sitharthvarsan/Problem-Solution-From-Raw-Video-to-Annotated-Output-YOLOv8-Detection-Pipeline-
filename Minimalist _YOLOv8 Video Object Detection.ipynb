```python
from ultralytics import YOLO
import cv2
import numpy as np
from IPython.display import display, Image, clear_output  # Import clear_output separately
import os

# Initialize YOLOv8 model
model=YOLO(r"C:\Users\HP\yolov8n.pt")  # or yolov8s.pt, yolov8m.pt

# Video paths
input_video_path = r"C:\Users\HP\OneDrive\Desktop\INTERN\d.mp4"  # Replace with your video path
output_video_path = r"C:\Users\HP\OneDrive\Desktop\INTERN\output_annotated.mp4"

# Open video capture
cap = cv2.VideoCapture(input_video_path)

# Get video properties
fps = int(cap.get(cv2.CAP_PROP_FPS))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Initialize video writer
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_video_path, fourcc, fps, (width, height))

# Process video frame-by-frame
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    
    # Run detection
    results = model(frame, conf=0.5)  # Adjust confidence as needed
    
    # Draw detections on frame
    annotated_frame = results[0].plot()  # Automatically adds bounding boxes
    
    # Write frame to output video
    out.write(annotated_frame)
    
    # Display frame in Jupyter (converts BGR to RGB)
    display(Image(data=cv2.imencode('.jpg', annotated_frame)[1].tobytes()))
    
    # Clear output to avoid stacking frames
    clear_output(wait=True)  # Use clear_output as a function, not a method

# Release resources
cap.release()
out.release()
print(f"Done! Output saved to {output_video_path}")
```
